```python
import pandas as pd
import numpy as np
import os, sys
```

```python
df = pd.read_csv(r'E:\Projects\network-security-mlops\data\Phishing_Legitimate_full.csv')
df.head()
```

```python
# the dataframe's schema about the columns informations
# show in the yaml file
# format:
    # columns:
    #   - id: int64
    #   - NumDots: int64
    # numerical_columns:
    #   - NumDots: int64
    # categorical_columns:
    #   - a_categorical_column: object

with open(r'E:\Projects\network-security-mlops\data_schema\schema.yaml', 'w') as f:
    f.write("columns:\n")
    for column in df.columns:
        f.write(f"  - {column}: {df[column].dtype}\n")
    f.write("numerical_columns:\n")
    for column in df.columns:
        if df[column].dtype == 'int64' or \
            df[column].dtype == 'float64' or \
            df[column].dtype == 'bool':
            f.write(f"  - {column}\n")
    f.write("categorical_columns:\n")
    for column in df.columns:
        if df[column].dtype == 'object' or df[column].dtype == 'category' or df[column].dtype == 'str':
            f.write(f"  - {column}\n")
```
